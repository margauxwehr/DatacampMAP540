```python
from bs4 import BeautifulSoup
import requests
import pandas
import re
import urllib
from urllib.request import urlopen
```

# create empty data frame with desired columns

```python
phone = pandas.DataFrame({'name' : [], 'review': []})
```

# load page and convert to a beautiful soup object 

```python
quote_page = ("https://www.bestbuy.com/site/reviews/apple-iphone-x-256gb-space-gray-at-t/6009664")
```

```python
result = []
numberpages = 10
for i in range(1,numberpages):
    url = "https://www.bestbuy.com/site/reviews/apple-iphone-x-256gb-space-gray-at-t/6009664?page=" + str(i)
    req = urllib.request.Request(url)
    page = urlopen(req)
    soup = BeautifulSoup(page, 'html.parser')
    comments = soup.select(".reviews-content-wrapper .review-item")

    for comment in comments:
        title = comment.select_one("h4").get_text()
        date = comment.select_one(".review-date").get_text()
        text = comment.select_one(".review-content").get_text()
        rating = int(comment.select_one(".reviewer-score").get_text())
        result.append({
            'title': title,
            'date': date,
            'text': text,
            'rating': rating
        })

pandas.DataFrame(result)
    

#<span class="BVRRReviewText">Well I must say I love my new iPhone 7 coming from a galaxy phone to an apple phone it very nice</span>
#<div class="BVRRRatingNormalImage BVRRRatingNormalImageCustomRating_5"><img src="https://t-mobile.ugc.bazaarvoice.com/9060redes2-en_us/5_0/5/rating.gif" class="BVImgOrSprite" alt="5 / 5" title="5 / 5"></div>
```
